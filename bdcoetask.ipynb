```python
total_confirmed = df["Confirmed"].sum()
total_deaths = df["Deaths"].sum()
total_recoveries = df["Recovered"].sum()
max_confirmed = df["Confirmed"].max()
min_confirmed = df["Confirmed"].min()
max_deaths = df["Deaths"].max()
min_deaths = df["Deaths"].min()
max_recoveries = df["Recovered"].max()
min_recoveries = df["Recovered"].min()
```

```python
print("Total Confirmed Cases:", total_confirmed)
print("Total Deaths:", total_deaths)
print("Total Recoveries:", total_recoveries)
print("Max Confirmed Cases:", max_confirmed)
print("Min Confirmed Cases:", min_confirmed)
print("Max Deaths:", max_deaths)
print("Min Deaths:", min_deaths)
print("Max Recoveries:", max_recoveries)
print("Min Recoveries:", min_recoveries)
```
